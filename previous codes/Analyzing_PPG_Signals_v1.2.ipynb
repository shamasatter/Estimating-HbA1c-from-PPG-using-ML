> __Algorithm 00:__ Split Transmitted and Reflected Signal from a distinct PPG Signal 
>> ACK: __Reconstructed by Shama__. _Previous Code is being studied and converted into a function for better utilization_

```python
def previousCode_split_transmitted_and_reflected_from_PPG (combinedSignal, selectIndex, drawFigure=False):
    # Initialization variables
    transmittedSignal = []
    reflectedSignal   = []

    # Check one by one data
    for i in range(0, len(combinedSignal)):
        if combinedSignal.iloc[i,-1] == 1 :
            # Inserting data one by one
            transmittedSignal = np.append(transmittedSignal, combinedSignal.iloc[i,selectIndex])
        elif combinedSignal.iloc[i,-1] == 0 or combinedSignal.iloc[i,-1] > 1 :
            # Inserting data one by one
            reflectedSignal   = np.append(reflectedSignal,   combinedSignal.iloc[i,selectIndex])

    # Deleting few data from the beginning
    transmittedSignal   = np.delete(transmittedSignal, 0, axis =0)
    reflectedSignal     = np.delete(reflectedSignal,   0, axis =0)
    
    # Visualization **Don't mention inside the pseudo code**
    if drawFigure:
        plt.rcParams.update(plt.rcParamsDefault)
        plt.rcParams['font.size'] = '8'
        fig,axs = plt.subplots(4, figsize=(20, 3))
        axs[0].set(title='Visualization of Transmitted and Reflected Signal')
        axs[0].plot(combinedSignal.iloc[:,-1], color='grey', label='T/R signal')
        axs[0].set(ylabel = ' ')
        axs[0].legend(loc=1)
        axs[0].grid(True)
        axs[1].plot(combinedSignal.iloc[:,selectIndex], color='orange', label='Selected signal')
        axs[1].set(ylabel = ' ', xlabel = ' ')
        axs[1].legend(loc=1)
        axs[1].grid(True)
        axs[2].plot(pd.DataFrame(transmittedSignal), color='red', label='Transmitted signal')
        axs[2].set(ylabel = ' ', xlabel = ' ')
        axs[2].legend(loc=1)
        axs[2].grid(True)
        axs[3].plot(pd.DataFrame(reflectedSignal), color='blue', label='Reflected signal')
        axs[3].set(ylabel = ' ', xlabel = ' ')
        axs[3].legend(loc=1)
        axs[3].grid(True)
        plt.show()
        
    # Return
    return transmittedSignal,reflectedSignal
```

> __Algorithm 01:__ Split Transmitted and Reflected Signal from a distinct PPG Signal 
>> ACK: __Developed by Shama__. _This algorithm significantly improves the performance of the trained models_

```python
def upgradedCode_split_transmitted_and_reflected_from_PPG (combinedSignal, selectIndex, removeNoise=0, drawFigure=False):
    # Initialization variables
    transmittedSignal = []
    reflectedSignal   = []
    CountData_t       = 0
    CountData_r       = 0

    # Check one by one data
    for i in range(0, len(combinedSignal)):
        if combinedSignal.iloc[i,-1] == 1 :
            CountData_r  = 0
            CountData_t += 1
            if CountData_t <= removeNoise:
                # Deleting last data one by one
                reflectedSignal   = reflectedSignal  [:-1]
            elif CountData_t > removeNoise:
                # Inserting data one by one
                transmittedSignal = np.append(transmittedSignal, combinedSignal.iloc[i,selectIndex])
        elif combinedSignal.iloc[i,-1] == 0 or combinedSignal.iloc[i,-1] > 1 :
            CountData_t  = 0
            CountData_r += 1
            if CountData_r <= removeNoise:
                # Deleting last data one by one
                transmittedSignal   = transmittedSignal  [:-1]
            elif CountData_r > removeNoise:
                # Inserting data one by one
                reflectedSignal   = np.append(reflectedSignal,   combinedSignal.iloc[i,selectIndex])

    # Deleting last few data based on counting removeNoise
    transmittedSignal = transmittedSignal[:-removeNoise]
    reflectedSignal   = reflectedSignal  [:-removeNoise]
    
    # Visualization **Don't mention inside the pseudo code**
    if drawFigure:
        plt.rcParams.update(plt.rcParamsDefault)
        plt.rcParams['font.size'] = '8'
        fig,axs = plt.subplots(4, figsize=(20, 3))
        axs[0].set(title='Visualization of Transmitted and Reflected Signal')
        axs[0].plot(combinedSignal.iloc[:,-1], color='grey', label='T/R signal')
        axs[0].set(ylabel = ' ')
        axs[0].legend(loc=1)
        axs[0].grid(True)
        axs[1].plot(combinedSignal.iloc[:,selectIndex], color='orange', label='Selected signal')
        axs[1].set(ylabel = ' ', xlabel = ' ')
        axs[1].legend(loc=1)
        axs[1].grid(True)
        axs[2].plot(pd.DataFrame(transmittedSignal), color='red', label='Transmitted signal')
        axs[2].set(ylabel = ' ', xlabel = ' ')
        axs[2].legend(loc=1)
        axs[2].grid(True)
        axs[3].plot(pd.DataFrame(reflectedSignal), color='blue', label='Reflected signal')
        axs[3].set(ylabel = ' ', xlabel = ' ')
        axs[3].legend(loc=1)
        axs[3].grid(True)
        plt.show()
        
    # Return
    return transmittedSignal,reflectedSignal
```

> __Algorithm 02:__ Baseline drift remove by utilizing windowing method
>> ACK: __Modified by Shama__. _a variable __comSignal__ is added to combine the splitted signal in order to determine better PPG signals_

```python
def baseline_drift_remove(rawSignal, drawFigure=False):
    # Initialization of variables
    comSignal   = []
    newSignal   = np.empty((150000, 200))
    windowSize  = 100
    y           = np.zeros((150000))
    x           = np.zeros((150000))
    # Making Frame
    signalFrame = librosa.util.frame(np.asarray(rawSignal), frame_length=windowSize, hop_length=windowSize, axis = 0)
    # Windowing Frame
    for i in range(0, signalFrame.shape[0]):
        y[i]  = np.min(signalFrame[i,:])
        index = np.argmin(signalFrame[i,:])
        x[i]  = i * windowSize + index
    # Removing baseline drift
    y = y[0:i]
    x = x[0:i]
    baseline_x = np.arange(0, rawSignal.shape[0])
    tck        = scipy.interpolate.splrep(x, y, s=0, k=1)
    baseline_y = scipy.interpolate.splev(baseline_x, tck)
    rawSignal  = rawSignal - baseline_y
    # Making multidimensional arrays
    for i in range(0, x.shape[0]):
        if (i == x.shape[0]-1):
            break
        sig_rs = rawSignal[int(x[i]):int(x[i]+200)].reshape(-1,1).T
        try:
            newSignal[i] = sig_rs
            comSignal    = np.append(comSignal,sig_rs)
        except:
            continue
    newSignal = newSignal[0:i]
    # Visualization **Don't mention inside the pseudo code**
    if drawFigure:
        plt.rcParams.update(plt.rcParamsDefault)
        plt.rcParams['font.size'] = '8'
        fig,axs = plt.subplots(2, figsize=(20, 2.5))
        axs[0].set(title='Visualization of PPG Signal before and after deploying BDR & Windowing')
        axs[0].plot(pd.DataFrame(rawSignal), color='red', label='Before BDR & Windowing')
        axs[0].set(ylabel = ' ', xlabel = ' ')
        axs[0].legend(loc=1)
        axs[0].grid(True)
        axs[1].plot(pd.DataFrame(comSignal), color='blue', label='After BDR & Windowing')
        axs[1].set(ylabel = ' ', xlabel = ' ')
        axs[1].legend(loc=1)
        axs[1].grid(True)
        plt.show()
    # Return
    return newSignal,comSignal
```
