> __Algorithm 01:__ Split Transmitted and Reflected Signal from a distinct PPG Signal 
>> ACK: __Developed by Shama__. _This algorithm significantly improves the performance of the trained models_

```python
def split_transmitted_and_reflected_from_PPG (combinedSignal, selectIndex, removeNoise=0, drawFigure=False):
    # Initialization variables
    transmittedSignal = []
    reflectedSignal   = []
    CountData_t       = 0
    CountData_r       = 0

    # Check one by one data
    for i in range(0, len(combinedSignal)):
        if combinedSignal.iloc[i,-1] == 1 :
            CountData_r  = 0
            CountData_t += 1
            if CountData_t <= removeNoise:
                # Deleting last data one by one
                reflectedSignal   = reflectedSignal  [:-1]
            elif CountData_t > removeNoise:
                # Inserting data one by one
                transmittedSignal = np.append(transmittedSignal, combinedSignal.iloc[i,selectIndex])
        elif combinedSignal.iloc[i,-1] == 0 or combinedSignal.iloc[i,-1] > 1 :
            CountData_t  = 0
            CountData_r += 1
            if CountData_r <= removeNoise:
                # Deleting last data one by one
                transmittedSignal   = transmittedSignal  [:-1]
            elif CountData_r > removeNoise:
                # Inserting data one by one
                reflectedSignal   = np.append(reflectedSignal,   combinedSignal.iloc[i,selectIndex])

    # Deleting last few data based on counting removeNoise
    transmittedSignal = transmittedSignal[:-removeNoise]
    reflectedSignal   = reflectedSignal  [:-removeNoise]
    
    # Visualization **Don't mention inside the pseudo code**
    if drawFigure:
        plt.rcParams.update(plt.rcParamsDefault)
        plt.rcParams['font.size'] = '12'
        fig,axs = plt.subplots(4, figsize=(20, 7))
        axs[0].set(title='Visualization of Transmitted and Reflected Signal')
        axs[0].plot(combinedSignal.iloc[:,-1], color='grey', label='T/R signal')
        axs[0].set(ylabel = ' ')
        axs[0].legend(loc=1)
        axs[0].grid(True)
        axs[1].plot(combinedSignal.iloc[:,selectIndex], color='orange', label='Selected signal')
        axs[1].set(ylabel = ' ', xlabel = ' ')
        axs[1].legend(loc=1)
        axs[1].grid(True)
        axs[2].plot(pd.DataFrame(transmittedSignal), color='red', label='Transmitted signal')
        axs[2].set(ylabel = ' ', xlabel = ' ')
        axs[2].legend(loc=1)
        axs[2].grid(True)
        axs[3].plot(pd.DataFrame(reflectedSignal), color='blue', label='Reflected signal')
        axs[3].set(ylabel = ' ', xlabel = ' ')
        axs[3].legend(loc=1)
        axs[3].grid(True)
        plt.show()
        
    # Return
    return transmittedSignal,reflectedSignal
```

> __Algorithm 02:__ Baseline drift remove by utilizing windowing method
>> ACK: __Modified by Shama__. _a variable __comSignal__ is added to combine the splitted signal in order to determine better PPG signals_

```python
def baseline_drift_remove(rawSignal, drawFigure=False):
    # Initialization of variables
    comSignal   = []
    newSignal   = np.empty((150000, 200))
    windowSize  = 100
    y           = np.zeros((150000))
    x           = np.zeros((150000))
    # Making Frame
    signalFrame = librosa.util.frame(np.asarray(rawSignal), frame_length=windowSize, hop_length=windowSize, axis = 0)
    # Windowing Frame
    for i in range(0, signalFrame.shape[0]):
        y[i]  = np.min(signalFrame[i,:])
        index = np.argmin(signalFrame[i,:])
        x[i]  = i * windowSize + index
    # Removing baseline drift
    y = y[0:i]
    x = x[0:i]
    baseline_x = np.arange(0, rawSignal.shape[0])
    tck        = scipy.interpolate.splrep(x, y, s=0, k=1)
    baseline_y = scipy.interpolate.splev(baseline_x, tck)
    rawSignal  = rawSignal - baseline_y
    # Making multidimensional arrays
    for i in range(0, x.shape[0]):
        if (i == x.shape[0]-1):
            break
        sig_rs = rawSignal[int(x[i]):int(x[i]+200)].reshape(-1,1).T
        try:
            newSignal[i] = sig_rs
            comSignal    = np.append(comSignal,sig_rs)
        except:
            continue
    newSignal = newSignal[0:i]
    # Visualization **Don't mention inside the pseudo code**
    if drawFigure:
        plt.rcParams.update(plt.rcParamsDefault)
        plt.rcParams['font.size'] = '12'
        fig,axs = plt.subplots(2, figsize=(20, 3.5))
        axs[0].set(title='Visualization of PPG Signal before and after deploying BDR & Windowing')
        axs[0].plot(pd.DataFrame(rawSignal), color='red', label='Before BDR & Windowing')
        axs[0].set(ylabel = ' ', xlabel = ' ')
        axs[0].legend(loc=1)
        axs[0].grid(True)
        axs[1].plot(pd.DataFrame(comSignal), color='blue', label='After BDR & Windowing')
        axs[1].set(ylabel = ' ', xlabel = ' ')
        axs[1].legend(loc=1)
        axs[1].grid(True)
        plt.show()
    # Return
    return newSignal,comSignal
```

> __Algorithm 03:__ Butter Low-pass Filter (Ignore this in SCI)

```python
def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = scipy.signal.butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_lowpass_filter(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = scipy.signal.lfilter(b, a, data)
    return y
```

> __Algorithm 04:__ Kaiser–Teager Energy

```python
def get_kaiser_teager_energy(x):
    x         = np.array(x)
    l         = len(x)
    squ       = x[1:l-1]**2
    oddi      = x[0:l-2]
    eveni     = x[2:l]
    ex        = squ - (oddi*eveni)
    kte       = np.zeros([l,])
    kte[0]    = ex[0]
    kte[-1]   = ex[-1]
    kte[1:-1] = ex
    return kte
```

> __Algorithm 05:__ Determine AC_DC_from_Signal
>> ACK: __Developed by Shama__. _This algorithm is developed to determine __r1__ and __r2__ which significantly improve the performance of the trained models_

```python
def determine_ac_dc_from_PPG (Signal):
    uPeak =  scipy.signal.find_peaks(Signal)
    uPeak =  pd.DataFrame(uPeak[0])
    lPeak =  scipy.signal.find_peaks(-Signal)
    lPeak =  pd.DataFrame(lPeak[0])
    
    if (len(uPeak)<len(lPeak)):
        indexLength = len(uPeak)
    else:
        indexLength = len(lPeak)

    for index in range(indexLength):
        newValue_ac =  Signal[uPeak.iloc[index]] - Signal[lPeak.iloc[index]]
        newValue_dc =  Signal[lPeak.iloc[index]]
        if index == 0:
            allValue_ac = newValue_ac
            allValue_dc = newValue_dc
        elif index > 0:
            allValue_ac =  np.append(allValue_ac,newValue_ac)
            allValue_dc =  np.append(allValue_dc,newValue_dc)

    meanValue_ac = np.mean(allValue_ac)
    meanValue_dc = np.mean(allValue_dc)

    return meanValue_ac,meanValue_dc
```

> __Function 01:__ BA_Plot

```python
def bland_altman_plot(data1, data2):
    data1     = np.asarray(data1)
    data2     = np.asarray(data2)
    mean      = np.mean([data1, data2], axis=0)
    diff      = data1 - data2                   # Difference between data1 and data2
    md        = np.mean(diff)                   # Mean of the difference
    sd        = np.std(diff, axis=0)            # Standard deviation of the difference

    plt.scatter(mean, diff,   c = 'black')
    plt.axhline(md,           color='green', linestyle='--')
    plt.axhline(md + 1.96*sd, color='red',   linestyle='--')
    plt.axhline(md - 1.96*sd, color='red',   linestyle='--')
    plt.xlabel("Average of the two measures", fontsize = 14)
    plt.ylabel("Difference between the two measures", fontsize = 14)
    plt.title("Bland-Altman plot for estimated %HbA1c values", fontsize = 14)
    # plt.savefig('./Figure/RF_trans_BA_Plot.png', bbox_inches = "tight", transparent = True, dpi = 400)
```

```python
def construct_feature_set_v5 (PPG, HbA1c, SpO2, FW, BMI, signalType='Transmitted', removeNoise=1, samplingRate=37):
    # Initialization
    sr         = samplingRate
    l          = 0

    for i in range (0,len(PPG)):
        
        # Getting corresponding Combined PPG Data 
        combinedSignal     = pd.read_csv(PPG[i], delimiter=',', header=None)
        print(str(PPG[i]))
       
        # __Algorithm 01:__ Split Transmitted and Reflected Signal from a distinct PPG Signal 
        transSignal_g, reflecSignal_g = split_transmitted_and_reflected_from_PPG(
            combinedSignal = combinedSignal, 
            selectIndex    = 1, #Green
            removeNoise    = removeNoise, 
            drawFigure     = False
            )
        transSignal_r, reflecSignal_r = split_transmitted_and_reflected_from_PPG(
            combinedSignal = combinedSignal, 
            selectIndex    = 2, #Red
            removeNoise    = removeNoise, 
            drawFigure     = False
            )
        transSignal_b, reflecSignal_b = split_transmitted_and_reflected_from_PPG(
            combinedSignal = combinedSignal, 
            selectIndex    = 0, #Blue
            removeNoise    = removeNoise, 
            drawFigure     = False
            )
        
        if signalType == 'Transmitted':
            # > __Algorithm 02:__ Baseline drift remove by utilizing windowing method
            bdrSignal_g, comS_g = baseline_drift_remove(
                rawSignal       = transSignal_g,  # Only Transmitted Signal
                drawFigure      = False
                )
            bdrSignal_r, comS_r = baseline_drift_remove(
                rawSignal       = transSignal_r,  # Only Transmitted Signal
                drawFigure      = False
                )
            bdrSignal_b, comS_b = baseline_drift_remove(
                rawSignal       = transSignal_b,  # Only Transmitted Signal
                drawFigure      = False
                )
            # > __Algorithm 05:__ Determine AC_DC_from_Signal
            r_ac,r_dc = determine_ac_dc_from_PPG(transSignal_r)
            g_ac,g_dc = determine_ac_dc_from_PPG(transSignal_g)
            b_ac,b_dc = determine_ac_dc_from_PPG(transSignal_b)

        elif signalType == 'Reflected':
            # > __Algorithm 02:__ Baseline drift remove by utilizing windowing method
            bdrSignal_g, comS_g = baseline_drift_remove(
                rawSignal       = reflecSignal_g,  # Only Reflected Signal
                drawFigure      = False
                )
            bdrSignal_r, comS_r = baseline_drift_remove(
                rawSignal       = reflecSignal_r,  # Only Reflected Signal
                drawFigure      = False
                )
            bdrSignal_b, comS_b = baseline_drift_remove(
                rawSignal       = reflecSignal_b,  # Only Reflected Signal
                drawFigure      = False
                )
            # > __Algorithm 05:__ Determine AC_DC_from_Signal
            r_ac,r_dc = determine_ac_dc_from_PPG(reflecSignal_r)
            g_ac,g_dc = determine_ac_dc_from_PPG(reflecSignal_g)
            b_ac,b_dc = determine_ac_dc_from_PPG(reflecSignal_b)


        # Calculate r1 & r2
        r1 = (g_ac/g_dc)/(r_ac/r_dc)
        r2 = (b_ac/b_dc)/(r_ac/r_dc)
        
        
        # Constructing Feature Set
        for j in range(len(bdrSignal_r)):

            featureFrame       = []
            
            # >>>>>>>>>>>>>>>>Red>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
            # > __Algorithm 03:__ Butter Low-pass Filter
            filteredSignal_r           = butter_lowpass_filter(
                data  = bdrSignal_r[j,:]-np.mean(bdrSignal_r[j,:]), 
                cutoff= 8, 
                fs    = sr, 
                order = 6
                )

            fqSignal_r, psdSignal_r    = scipy.signal.welch(
                x       = filteredSignal_r, 
                nperseg = len(filteredSignal_r) # Added to overcome error
                ) 


            # > __Algorithm 04:__ Kaiser–Teager Energy
            kteSignal_r            = get_kaiser_teager_energy(
                filteredSignal_r
                )



            # Insert Feature one by one
            # Add new feature (5)
            ar_coeffs_r, _r        = statsmodels.regression.linear_model.yule_walker(filteredSignal_r, order = 5)
            # featureFrame           = np.append(featureFrame,ar_coeffs_r)
            
            # Add new feature
            zcSignal_r             = tsfel.feature_extraction.features.zero_cross(filteredSignal_r)
            featureFrame           = np.append(featureFrame,zcSignal_r)
            
            # Add new feature
            skewSignal_r           = tsfel.feature_extraction.features.skewness(filteredSignal_r)
            # featureFrame           = np.append(featureFrame,skewSignal_r)
            
            # Add new feature
            absWavelength_r        = tsfel.feature_extraction.wavelet_abs_mean(filteredSignal_r)
            maWavelength_r         = np.mean(absWavelength_r)
            # featureFrame           = np.append(featureFrame,maWavelength_r)
            
            # Add new feature
            autocorr_r             = tsfel.feature_extraction.autocorr(filteredSignal_r)
            # featureFrame           = np.append(featureFrame,autocorr_r)
            
            # Add new feature
            kurtSpectral_r         = tsfel.feature_extraction.features.spectral_kurtosis(filteredSignal_r, sr)
            # featureFrame           = np.append(featureFrame,kurtSpectral_r)
            
            # Add new feature
            skewSpectral_r         = tsfel.feature_extraction.features.spectral_skewness(filteredSignal_r, sr)
            # featureFrame           = np.append(featureFrame,skewSpectral_r)
            
            # Add new feature
            sum_abs_diff_r         = tsfel.feature_extraction.features.sum_abs_diff(filteredSignal_r)
            featureFrame           = np.append(featureFrame,sum_abs_diff_r)
            
            # Add new feature
            kurtPSD_r              = scipy.stats.kurtosis(psdSignal_r)
            # featureFrame           = np.append(featureFrame,kurtPSD_r)
            
            # Add new feature
            varPSD_r               = np.var(psdSignal_r)
            featureFrame           = np.append(featureFrame,varPSD_r)
            
            # Add new feature
            meanPSD_r              = np.mean(psdSignal_r)
            # featureFrame           = np.append(featureFrame,meanPSD_r)
            
            # Add new feature
            skewKTE_r              = scipy.stats.skew(kteSignal_r)
            # featureFrame           = np.append(featureFrame,skewKTE_r)
            
            # Add new feature
            kurtKTE_r              = scipy.stats.kurtosis(kteSignal_r)
            # featureFrame           = np.append(featureFrame,kurtKTE_r)
            
            # Add new feature
            varKTE_r               = np.var(kteSignal_r)
            featureFrame           = np.append(featureFrame,varKTE_r)
            
            # Add new feature
            meanKTE_r              = np.mean(kteSignal_r)
            # featureFrame           = np.append(featureFrame,meanKTE_r)
            
            
            # >>>>>>>>>>>>>>>>Green>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
           
            # > __Algorithm 03:__ Butter Low-pass Filter
            filteredSignal_g           = butter_lowpass_filter(
                data  = bdrSignal_g[j,:]-np.mean(bdrSignal_g[j,:]), 
                cutoff= 8, 
                fs    = sr, 
                order = 6
                )
           
            fqSignal_g, psdSignal_g    = scipy.signal.welch(
                x       = filteredSignal_g, 
                nperseg = len(filteredSignal_g) # Added to overcome error
                ) 
           
            # > __Algorithm 04:__ Kaiser–Teager Energy
            kteSignal_g            = get_kaiser_teager_energy(
                filteredSignal_g
                )
            
            
            # Insert Feature one by one
            # Add new feature
            ar_coeffs_g, _g        = statsmodels.regression.linear_model.yule_walker(filteredSignal_g, order = 5)
            # featureFrame           = np.append(featureFrame,ar_coeffs_g)
            
            # Add new feature
            zcSignal_g             = tsfel.feature_extraction.features.zero_cross(filteredSignal_g)
            featureFrame           = np.append(featureFrame,zcSignal_g)
            
            # Add new feature
            skewSignal_g           = tsfel.feature_extraction.features.skewness(filteredSignal_g)
            # featureFrame           = np.append(featureFrame,skewSignal_g)
            
            # Add new feature
            absWavelength_g        = tsfel.feature_extraction.wavelet_abs_mean(filteredSignal_g)
            maWavelength_g         = np.mean(absWavelength_g)
            # featureFrame           = np.append(featureFrame,maWavelength_g)
            
            # Add new feature
            autocorr_g             = tsfel.feature_extraction.autocorr(filteredSignal_g)
            # featureFrame           = np.append(featureFrame,autocorr_g)
            
            # Add new feature
            kurtSpectral_g         = tsfel.feature_extraction.features.spectral_kurtosis(filteredSignal_g, sr)
            # featureFrame           = np.append(featureFrame,kurtSpectral_g)
            
            # Add new feature
            skewSpectral_g         = tsfel.feature_extraction.features.spectral_skewness(filteredSignal_g, sr)
            # featureFrame           = np.append(featureFrame,skewSpectral_g)
            
            # Add new feature
            sum_abs_diff_g         = tsfel.feature_extraction.features.sum_abs_diff(filteredSignal_g)
            featureFrame           = np.append(featureFrame,sum_abs_diff_g)
            
            # Add new feature
            kurtPSD_g              = scipy.stats.kurtosis(psdSignal_g)
            # featureFrame           = np.append(featureFrame,kurtPSD_g)
            
            # Add new feature
            varPSD_g               = np.var(psdSignal_g)
            featureFrame           = np.append(featureFrame,varPSD_g)
            
            # Add new feature
            meanPSD_g              = np.mean(psdSignal_g)
            # featureFrame           = np.append(featureFrame,meanPSD_g)
            
            # Add new feature
            skewKTE_g              = scipy.stats.skew(kteSignal_g)
            # featureFrame           = np.append(featureFrame,skewKTE_g)
            
            # Add new feature
            kurtKTE_g              = scipy.stats.kurtosis(kteSignal_g)
            # featureFrame           = np.append(featureFrame,kurtKTE_g)
            
            # Add new feature
            varKTE_g               = np.var(kteSignal_g)
            featureFrame           = np.append(featureFrame,varKTE_g)
            
            # Add new feature
            meanKTE_g              = np.mean(kteSignal_g)
            # featureFrame           = np.append(featureFrame,meanKTE_g)


            # >>>>>>>>>>>>>>>>Add_r1_r2_as_features>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
            featureFrame           = np.append(featureFrame,r1)
            featureFrame           = np.append(featureFrame,r2)
            
            
            # >>>>>>>>>>>>>>>>Add_other_features>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
            # featureFrame           = np.append(featureFrame,SpO2[i])
            # featureFrame           = np.append(featureFrame,FW[i])
            # featureFrame           = np.append(featureFrame,BMI[i])

            # >>>>>>>>>>>>>>> Add output feature 
            featureFrame           = np.append(featureFrame,HbA1c[i])

            ## Insert featureFrame into featureSet
            # Initialized feature set array 
            if l == 0:
                k = len(featureFrame)
                # print('kCount='+str(k))
                featureSet = np.empty([150000,k])
            # Save into featureSet
            featureSet[l][:k] = featureFrame
            # Length increased for next cycle
            l = l + 1


            # Checking outputs
            print('jCount='+str(j)+'||'+'lCount='+str(l)+'||'+'fCount='+str(len(featureFrame)))
            print(featureFrame)

        # Check point    
        # print('iCount='+str(i))

    # Gather all
    print('**********************************************************************')
    print('*****************************Summary**********************************')    
    featureSet = featureSet[0:l][:]
    # Check for null
    featureSet = featureSet[~np.isnan(featureSet).any(axis=1)]
    print('**********************************************************************')
    return featureSet
```

```python
# get a stacking ensemble of models
def get_stacking():
	# define the base models
	level0 = list()
	level0.append(('rf', RandomForestRegressorRegressor()))
	level0.append(('xg', XGBRegressor()))
	level0.append(('cb', CatBoostRegressor()))
	level0.append(('lg', LGBMRegressor()))
	# define meta learner model
	level1 = XGBRegressor()
	# define the stacking ensemble
	model = StackingRegressor(estimators=level0, final_estimator=level1, cv=5)
	return model
```
